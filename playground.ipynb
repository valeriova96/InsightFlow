```python
import sklearn

data = sklearn.datasets.load_breast_cancer(as_frame=True)
data
```

```python
import pandas as pd

my_df = pd.DataFrame(data.data, columns=data.feature_names)
my_df['target'] = data.target
my_df['target'] = my_df['target'].map({0: 'malignant', 1: 'benign'})
my_df.to_csv('breast_cancer.csv', index=False)
```
